```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import joblib
```

```python


df2 = pd.read_csv("/Users/swetarai/BLOOD_SYNOVIUM_COMBINE/GPL570_2_DATA_BLOOD_SYNO/272_776_GPL570_BLOOD_SYNO_exp_pheno.csv", index_col=0)
df2
```

```python
df2.columns
```

```python
df2= df2.drop("Gene_Symbol", axis=1)
df2
```

## label encoding

```python

# Label Encoding
label_encoder = LabelEncoder()
df2['gender'] = label_encoder.fit_transform(df2['gender'])
df2['sample_type'] = label_encoder.fit_transform(df2['sample_type'])
df2['disease_state'] = label_encoder.fit_transform(df2['disease_state'])

df2
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df2.drop(['gender', 'sample_type','disease_state'], axis=1),
    df2['disease_state'], # target
    test_size=0.2, 
    random_state=42,
)

X_train.shape, X_test.shape
```

```python
# create scaler
scaler = MinMaxScaler()
scaler.fit(X_train) 
X_train = pd.DataFrame(
    scaler.transform(X_train),
    columns=X_train.columns
)

X_test = pd.DataFrame(
    scaler.transform(X_test),
    columns=X_train.columns
)
```

# save the trian and test data

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)

y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
