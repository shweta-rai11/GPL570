```python
import pandas as pd
```

```python
M1 = pd.read_csv("/Users/swetarai/BLOOD_SYNOVIUM_COMBINE/gse_272_blood/annotated_exp_272.csv")
M1["Gene Symbol"].isnull().sum()
df_1 = M1[M1["Gene Symbol"].notnull()]
filtered_df_1 = df_1.loc[:, ['Gene Symbol'] + [col for col in df_1.columns if col.startswith("GSM")]]
filtered_df_1 = filtered_df_1.drop_duplicates(subset='Gene Symbol', keep='first')
filtered_df_1
```

```python
M2 = pd.read_csv("/Users/swetarai/BLOOD_SYNOVIUM_COMBINE/gse_776_blood/annotated_exp_776.csv")
M2["Gene Symbol"].isnull().sum()
df_2 = M2[M2["Gene Symbol"].notnull()]
filtered_df_2 = df_2.loc[:, ['Gene Symbol'] + [col for col in df_2.columns if col.startswith("GSM")]]
filtered_df_2 = filtered_df_2.drop_duplicates(subset='Gene Symbol', keep='first')
filtered_df_2
```

```python
merged_data = pd.merge(filtered_df_1, filtered_df_2, on="Gene Symbol", how ='inner')
merged_data = merged_data.rename(columns={'Gene Symbol': 'Gene_Symbol'})
merged_data
```

```python
merged_data= merged_data.set_index("Gene_Symbol").T
```

```python
merged_data = merged_data.reset_index()
print(merged_data.columns)
merged_data = merged_data.rename(columns={'index': 'Gene_Symbol'})
```
